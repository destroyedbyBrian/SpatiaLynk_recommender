```python
import folium
import pandas as pd
from folium import Choropleth, Circle, Marker
from folium.plugins import HeatMap, MarkerCluster
import math
```

```python
poi_data = pd.read_csv("./POI_geopanda.csv")
print(poi_data.head())
```

### Map 1 : Hundred POI Markers

```python
m = folium.Map(
    location=[1.3521, 103.8198],  
    zoom_start=12,                
    tiles="cartodb positron"
)

for idx, row in poi_data.head(100).iterrows():
	Marker([row['lat'], row['lon']]).add_to(m)

m
```

```python
m.save("100POI-markers.html")
```

### Map 2 : Cluster Map

```python
m_2 = folium.Map(
    location=[1.3521, 103.8198],  
    zoom_start=12,                
    tiles="cartodb positron"
)

mc = MarkerCluster()
for idx, row in poi_data.iterrows():
    if not math.isnan(row['lon']) and not math.isnan(row['lat']):
        mc.add_child(Marker([row['lat'], row['lon']]))
m_2.add_child(mc)
```

```python
m_2.save("Cluster-map.html")
```

### Step 3 : Heat Map

```python
m_3 = folium.Map(
    location=[1.3521, 103.8198],  
    zoom_start=12,                
    tiles="cartodb positron"
)

HeatMap(data=poi_data[['lat', 'lon']], radius=10).add_to(m_3)
```

```python
m_3.save("Heat-map.html")
```
